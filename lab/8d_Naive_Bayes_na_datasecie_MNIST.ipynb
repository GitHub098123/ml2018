Zaimplementować klasy `MultinomialNaiveBayes` oraz `BernoulliNaiveBayes`. Przetestować na datasecie MNIST.

Różnice w stosunku do Zadania 6:
* `X` jest tablicą (gęstą lub rzadką) o rozmiarze (liczba_obserwacji, liczba_cech),
* nie konwertujemy `X` wewnątrz implementowanych klas,
* brak atrybutu `features`,
* metoda `generate` zwraca jeden wygenerowany wiersz, czyli tablicę o rozmiarze (liczba_cech,).

Trzeba też zastanowić się, jak sensownie zaimplementować metodę `generate` w wersji Multinomial.

```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np
from sklearn.datasets import fetch_mldata

from zadania import BernoulliNaiveBayes, MultinomialNaiveBayes

mnist = fetch_mldata('MNIST original')
X = mnist.data
y = mnist.target.astype(np.int)
```

```python
bnb = BernoulliNaiveBayes(X, y)
mnb = MultinomialNaiveBayes(X, y)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(10):
    digit = bnb.likelihoods[i].reshape(28,28)
    fig.add_subplot(3, 4, i+1).imshow(digit, cmap="gray")
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(10):
    digit = bnb.generate(i).reshape(28,28)
    fig.add_subplot(3, 4, i+1).imshow(digit, cmap="gray")
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(10):
    digit = mnb.likelihoods[i].reshape(28,28)
    fig.add_subplot(3, 4, i+1).imshow(digit, cmap="gray")
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(10):
    digit = mnb.generate(i,20).reshape(28,28)
    fig.add_subplot(10, 5, 5*i+1).imshow(digit, cmap="gray")
    digit = mnb.generate(i,100).reshape(28,28)
    fig.add_subplot(10, 5, 5*i+2).imshow(digit, cmap="gray")
    digit = mnb.generate(i,784).reshape(28,28)
    fig.add_subplot(10, 5, 5*i+3).imshow(digit, cmap="gray")
    digit = mnb.generate(i,1000).reshape(28,28)
    fig.add_subplot(10, 5, 5*i+4).imshow(digit, cmap="gray")
    digit = mnb.generate(i,3000).reshape(28,28)
    fig.add_subplot(10, 5, 5*i+5).imshow(digit, cmap="gray")
```
