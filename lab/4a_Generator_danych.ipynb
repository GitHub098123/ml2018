```python
import matplotlib
%matplotlib notebook
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

from lab4 import ToyDataGenerator
```

```python
seed = 437
sigma = .3 # .03, .3, 3.
degree = 12 # 2, 6, 12, 100
n_samples = 50 # 50, 500, 5000, 50000

tdg = ToyDataGenerator(seed=seed, sigma=sigma, degree=degree)
X, Y = tdg.sample(n_samples)
X_noiseless, Y_noiseless = tdg.noiseless_sample(3000)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(X_noiseless, Y_noiseless, alpha=.6, label="noiseless", s=5)
ax.scatter(X, Y, alpha=.6, label="observed")
ax.legend()
```
