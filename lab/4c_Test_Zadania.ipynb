```python
import matplotlib
%matplotlib notebook
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

import numpy as np

from zadania import bias_variance_estimation
```

```python
def plot_bias_variance_estimation(degree, squared_bias, variance, noise, average_mse):
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)

    def colors_gen(n):
        cmap = matplotlib.cm.get_cmap('Vega20b')
        for i in np.linspace(0., 1., n):
            yield cmap(i)

    colors = colors_gen(5)

    ax.plot(degree, squared_bias, alpha=.6, label="squared_bias", c=next(colors))
    ax.plot(degree, variance, alpha=.6, label="variance", c=next(colors))
    ax.plot(degree, noise, alpha=.6, label="noise", c=next(colors))
    ax.plot(degree, squared_bias + variance + noise, alpha=.6, label="squared_bias + variance + noise", lw=15, c=next(colors))
    ax.plot(degree, average_mse, alpha=.6, label="mse", c=next(colors))
    ax.legend()
```

```python
degree, squared_bias, variance, noise, average_mse = bias_variance_estimation(
    generator_true_degree=3,
    generator_sigma=.3,
    generator_seed=437,
    n_train_samples=10,
    max_degree=6,
    n_simulations_per_degree=5)
```

```python
plot_bias_variance_estimation(degree, squared_bias, variance, noise, average_mse)
```
